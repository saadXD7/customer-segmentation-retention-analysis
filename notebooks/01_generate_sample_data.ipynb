# Generate Sample Customer Data
This notebook creates a realistic customer dataset for segmentation and retention analysis.

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random

# Set random seed for reproducibility
np.random.seed(42)
random.seed(42)

# Generate customer data
n_customers = 1000

data = {
    'customer_id': range(1, n_customers + 1),
    'signup_date': [datetime(2023, 1, 1) + timedelta(days=random.randint(0, 365)) for _ in range(n_customers)],
    'age': np.random.randint(18, 80, n_customers),
    'country': np.random.choice(['USA', 'UK', 'Canada', 'Germany', 'France', 'India'], n_customers),
    'total_purchases': np.random.randint(1, 50, n_customers),
    'total_spent': np.random.randint(100, 5000, n_customers),
    'last_purchase_days_ago': np.random.randint(1, 365, n_customers),
    'avg_order_value': np.random.randint(50, 500, n_customers),
    'is_active': np.random.choice([0, 1], n_customers, p=[0.3, 0.7]),
}

df = pd.DataFrame(data)

# Display first few rows
print("Sample Customer Data:")
print(df.head(10))
print(f"\nDataset Shape: {df.shape}")
print(f"\nColumn Info:\n{df.info()}")
print(f"\nBasic Statistics:\n{df.describe()}")
```

```python
# Save to CSV
df.to_csv('../data/raw/customer_data.csv', index=False)
print("Data saved to: data/raw/customer_data.csv")
```
